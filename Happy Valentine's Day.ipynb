```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (20,15)
import pylab
import scipy
```

```python
x = scipy.linspace(-2,2,1000)
y1 = scipy.sqrt(1-(abs(x)-1)**2)
y2 = -3*scipy.sqrt(1-(abs(x)/2)**0.5)
pylab.fill_between(x, y1, color='#F73D71')
pylab.fill_between(x, y2, color='#F73D71')
pylab.xlim([-2.5, 2.5])
#pylab.text(0, -0.4, '2021', fontsize=60, fontweight='bold',
           #color='white', horizontalalignment='center')
#pylab.savefig('heart.png')
```

```python
def fn(x,y,z):
    return ((x**2) + 9*(y**2)/4 + (z**2) - 1)**3-(x**2)*(z**3)-9*(y**2)*(z**3)/80
bbox=(-1.2,1.2)
xmin, xmax, ymin, ymax, zmin, zmax = bbox*3
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
A = np.linspace(xmin, xmax, 100)
B = np.linspace(xmin, xmax, 50)
A1,A2 = np.meshgrid(A,A)

for z in B:
    X,Y = A1,A2
    Z = fn(X,Y,z)
    cset = ax.contour(X, Y, Z+z, [z], zdir='z',colors='#F73D71')

for y in B:
    X,Z = A1,A2
    Y = fn(X,y,Z)
    cset = ax.contour(X, Y+y, Z, [y], zdir='y',colors='#FF95B3')

for x in B:
    Y,Z = A1,A2
    X = fn(x,Y,Z)
    cset = ax.contour(X+x, Y, Z, [x], zdir='x',colors='#ff668c')

ax.set_zlim3d(zmin,zmax)
ax.set_xlim3d(xmin,xmax)
ax.set_ylim3d(ymin,ymax)

plt.show()
```
